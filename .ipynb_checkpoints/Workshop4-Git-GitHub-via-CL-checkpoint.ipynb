## Tracking files "locally"

### Setting up git-tracking for a directory 📁
- Say you have a project that you have been developing code for (on your personal computer or hosted on some server like Haverford's JupyterHub or SciServer), and you would like to both save the current version of your code and track further changes to it. To do this, we will be using ```Git,``` which is an open source program for tracking changes to one or more text files. 
- To set up file tracking for a specific directory in the terminal, you should make sure you are in the directory and then type ```git init```
    - this will initialize the current directory as a git-tracked repository (repo)
    - if you list the full contents of the directory using ```ll -t``` you should see a new subdirectory called ```.git/``` has been created. This is where git stores all of the information needed to track your repository's file history.
- Before you can start tracking changes to your files, you must configure your author identity. This is important so that every commit you make to your local git-tracked repository will carry your identity and will be linked correctly when pushed to GitHub. This is especially important if multiple authors are making commits to the same repo. To do this, we will set our configuration settings to have our name and email. Type the following into the command line:
    - ```git config --global user.name "[your username]"``` (e.g., I would do ```git config --global user.name "ksanderso1"``` You can also use your real name if you'd like!)
    - ```git config --global user.email "[email used for GitHub account]"``` (e.g., I did ```git config --global user.email "ksanderso1@haverford.edu"```
    - Note: the --global flag tells git (the software) to use these same configurations for this and all future git tracked repositories, so you will only need to do this step once. 
    - Type ```git config --global --list``` to see these settings

### Tracking the files 🐾
- Initializing a git-tracked repo tells git that you would like this repository to be tracked, but it does not tell git which files in the repository to track. You have to option to tell git which files it should track. 
- Typing ```git status``` will give you information about the repository branch you are currently working on, any files that are not being tracked, and any files that you have made edits to, which can be committed (let's you view your staging area). 
- Let's tell git to track everything in the current directory (including files contained within sub-directories):
    - ```git add .``` (if you only want to track specific files, you would type ```git add file_name"```)
    - doing this stages the files for the first commit.
    - if you realize that you do not actually want to track all of the files, you can reset this actions using ```git reset .``` or ```git reset file_name```
- Type ```git status``` again. This will show you that the repository has new edits (adding files) which are staged and can be committed to the repo history
- Let's actually perform this commit, and save this version of the files in your git tracking history.
    - ```git commit -m "First commit: adding files"```
    - the ```-m``` flag is to attach a message
    - the stuff between the quotation marks is your commit message that summarizes the changes
    - The terminal will print out a summary of this commit along with your commit message
    - to undo the commit you can type ```git update-ref -d HEAD```
        - update-ref: manually update a pointer that Git uses internally to track commits.
        - d : delete flag means delete the reference
        - HEAD: a pointer to your current position in the repository history (most recent commit)
- After making this commit, you can type ```git status``` again to see a message telling you which branch you are working on, that you have nothing to commit ("working tree clean")
- To see the full version history of your repo, type ```git log```. 
    - This shows:
        - the commit ID
        - the author of the commit
        - a time stamp for the commit
        - the commit message
    - Later in the tutorial, we will learn how to use the commit ID to revert changes to a previous version

### Set up a connection between your local repo and the remote one you just created 📡
- Go back to your terminal
- First, let's rename our initial working branch ('master') to match the initial branch created for the remote repo ('main') we want to connect with. If your initial branch is already named 'main' you can skip this step. 
    - ```git branch -M main```
    - type ```git status``` to check that this worked.
- Create a new remote connection and name it origin (this is standard naming)
    - ```git remote add origin https://github.com/REPLACE WITH YOUR USERNAME/Workshop4-GitHub-practice-2026.git``` 
        - ```git remote -v``` will show you your remote connection is now set up
        - This shows that the remote connection 'origin' will be used for pushes (sending commits to github) and fetches (fetching commits/updates from GitHub that e.g., someone else has made to the repo)
- Upload the local main branch to the origin (remote) main branch 
    - ```git push -u origin main```
        - ```git push``` means upload the current version of the repo
        - ```origin``` is the remote GitHub repo
        - ```main``` is your local working branch
        - the ```-u``` flag sets origin as the default upstream branch for future pushes and pulls (so later pushes and pulls to this repo do not have to specify where to push/pull from)
- At this point, git will ask you for a username and password. Enter your GitHub username and the PAT we just generated.
- Go back to your GitHub account and refresh the page for the repo you just created. You should see the files appear.

### Let’s walk through a full workflow on the command line:

<p style="text-align: center;">branch → edit → stage → commit → push → (merge*) → repeat</p>


### First, a recap on terminology:
- **branch**: The branch of the repo that you are making changes on. 
    - By default, ```git init``` will create a single branch called 'main' or 'master'
    - it is good practice to do heavy development on a separate branch (e.g., 'dev')
- **stage**: the staging area is where git-tracked file changes are stored before their changes are committed to the repo history. 
    - This is where local git-tracking of file changes differs the most from the GitHub browser version control process.
    - In the browser, you make changes to a file and saving them automatically sets up a commit. Using git through the command line adds in this extra step that gives you space to prepare (stage) the changes that will be reflected in the next commit.
- **commit**: the process of making edits (adding, changing, deleting files) to the repository branch 
- **push**: the process of pushing local edits to the remote repo
- **merge**: the process of mergins edits from one branch with another
- **pull**: request to merge changes from the remote repo with a local branch. 

*merging is optional in this workflow. You do not need to merge the development branch with the main branch after every commit. It may make more sense to merge the development branch after a few commits that fix some bug or add some new feature.

### Create a branch for developing your code, and make some edits

#### Branch:
- ```git branch dev``` will create the new branch
- ```git checkout dev``` will place you on the dev branch
    - before making edits to any files, it is good practice to do ```git branch``` to verify which branch you are on
    
#### Edit:
- Go back to JupyterHub launch page and open the galaxies_analysis.py script. 
- In the print_stats() function, add a print statement for the mean of the input data frame. Make sure to edit the function’s docstring to include this information. Save your edits.
    - Test the code by running python3 galaxies_analysis.py from the command line
- Go back to terminal: ```git status``` will show that the file has been edited, but that this edit is not staged for commit

#### Stage:
- To stage the change type ```git add .``` then run ```git status``` again. The changed file will be highlighted in green.
    - this adds the change to the staging aread
    
#### Commit:
- To commit the change to the repo history type ```git commit -m "updating print_stats() function to include mean calculation”```
    - A summary of the commit will be printed telling you one file has been changed and the number of insertions and deletions
- Inspect the file changes made using: ```git diff main..dev -- galaxies_analysis.py```
    - ```git diff```: compare two branches
    - ```main..dev```: which two branches
    - ```-- galaxies_analysis.py```: only compare changes for this file
    
#### Push:
- Push the new branch to github’s remote repo using ```git push -u origin dev``` 
    - Enter username and PAT
    - Refresh GitHub repo (in the browser) to see the new branch appear
    - Notice that if you switch to the main branch in the GitHub browser, the changes have not yet appeared. This is because dev has not been merged with main. 

#### Merge:
- You can either merge this dev branch with main through the command line or through the browser like we did last time…let’s practice through the command line
    - Switch back to the main branch: ```git checkout main```
    - update main with dev changes (merge them): ```git merge dev```
    - Push the updated local main branch to the remote repo: ```git push```
        - Enter username and PAT
    - Go back to the remote repo on github and refresh the page to see the commits that were merged from dev to main
